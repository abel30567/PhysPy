```python
x = np.array([1,2])
xer = np.array([0.1,0.2])
y = np.array([2.2])
yer = np.array([0.15])
```

```python
xnew = unumpy.uarray(x, xer)
ynew = unumpy.uarray(y, yer)
```

```python
z = 4*xnew+5*ynew
print(unumpy.sqrt(z))
```

```python
a = np.array([1,2])
zval = unumpy.nominal_values(z)
zerr = unumpy.std_devs(z)
print(zval, np.sqrt(zval))
```

```python
plt.errorbar(a,zval,zerr, fmt ='o')
slope, intercept = np.polyfit(a, zval, deg=1)
print(slope, intercept)
plt.plot(a, intercept + slope*a)
plt.show()
```

```python
from uncertainties import unumpy as unp

distance = unp.uarray([6.0,7.0,8.0,9.3,10.7,11.9,12.8,13.8],[0.5,0.5,0.5,0.5,0.5,0.5,0.5,0.5])
counts = unp.uarray([211,138,126,103,78,66,53,36],[14.5,11.7,11.2,10.1,8.8,8.1,7.3,6.0])
countsnew = 1/unp.sqrt(counts)
```

```python
yvalues = unp.nominal_values(countsnew) # Dont forget square root
xvalues = unp.nominal_values(distance)

print(xvalues,
      yvalues)
```

```python
y_std_dev = unp.std_devs(countsnew)  
x_std_dev = unp.std_devs(distance)

# Plot values and errors
plt.errorbar(x = xvalues, y = yvalues, yerr = y_std_dev, fmt = 'o')
plt.title("Count vs distance")
plt.xlabel("xvals nd $\sigma_x$")
plt.ylabel("yvals")

#fit a linear line to the data
slope, intercept = np.polyfit(x = xvalues, y = yvalues, deg=1)
plt.plot(xvalues, slope*xvalues + intercept)

# Plot it all
plt.show()
print(slope)
```

```python
Sr_activity = ( 16 / (3.7 * (10**10) * (4.9**2) ) ) * ( counts * (distance**2) )
print(Sr_activity)
```

```python
Sr = Sr_activity.mean()
```

```python
print(unp.nominal_values(Sr), unp.std_devs(Sr))
```
